## Realizar una gráfica que permita ver el porcentaje de municipios por estado con índices de marginación "muy bajo", "bajo", "medio", "alto" y "muy alto". Guarda la gráfica en archivo png.

Agrupar por estado y contar los valores de la columna "Grado de marginacion"

```python
df_marginacion_estado = df_marginacion_clean.groupby(['Nombre de la entidad'], as_index=False)['Grado de marginación, 2020'].value_counts()
df_marginacion_estado
```

Se obtiene el porcentaje de marginacion por estado sumando el grado de marginacion respectivo de cada municipio por entidad federativa y dividiendolo por el total de grados de la entidad federativa.

```python
df_marginacion_estado['% de marginacion'] = 100 * df_marginacion_estado['count'] / df_marginacion_estado.groupby(['Nombre de la entidad'])['count'].transform('sum')
```

```python
df_marginacion_estado.sort_values('count', ascending=False)
```

```python
plt_scatter_marg_estado = px.scatter(df_marginacion_estado, 
                                     x='Nombre de la entidad',
                         y='% de marginacion',
                         color='Grado de marginación, 2020',
                         title='Porcentaje de municipios por estado con índices de marginación "muy bajo", "bajo", "medio", "alto" y "muy alto"',
                         hover_name='Nombre de la entidad',
                         )

plt_scatter_marg_estado.update_layout(xaxis_title = 'Numero de estados con un grado de marginacion',
                    yaxis_title = '% de grado de marginacion con respecto al total.')
plt_scatter_marg_estado.show()
```

```python
# Guardar grafica en formato png
plt_scatter_marg_estado.write_image("plt_scatter_marg_estado.png")
```

```python
plt_scatter_marg_estado_2 = px.scatter(df_marginacion_estado, 
                                     x='count',
                         y='% de marginacion',
                         color='Grado de marginación, 2020',
                         title='Porcentaje de municipios por estado con índices de marginación "muy bajo", "bajo", "medio", "alto" y "muy alto"',
                         hover_name='Nombre de la entidad',
                         )

plt_scatter_marg_estado_2.update_layout(xaxis_title = 'Numero de estados con un grado de marginacion',
                    yaxis_title = '% de grado de marginacion con respecto al total.')
plt_scatter_marg_estado_2.show()
```

## Realizar una gráfica que muestre el porcentaje de la población, respecto a la población total de cada estado, con índices de marginación "muy bajo", "bajo", "medio", "alto" y "muy alto". Guarda la gráfica en formato jpg.

```python
df_marginacion_clean['% poblacion total'] = 100 * df_marginacion_clean['Poblacion total'] / df_marginacion_clean.groupby('Nombre de la entidad')['Poblacion total'].transform('sum')
```

```python
df_marginacion_clean.head()
```

```python
plt_scatter = px.scatter(x=df_marginacion_clean['Poblacion total'],
                         y=df_marginacion_clean['% poblacion total'],
                         color=df_marginacion_clean['Grado de marginación, 2020'],
                         title='Porcentaje de la población, respecto a la población total de cada estado y su grado de marginación',
                         hover_name=df_marginacion_clean['Nombre de la entidad'],
                         )

plt_scatter.update_layout(xaxis_title = 'Poblacion total',
                    yaxis_title = '% poblacion total')
plt_scatter.show()
```

```python
# Guardar grafica en formato jpg
plt_scatter.write_image("plt_scatter_pobl_respecto_est.jpg")
```

## ¿Hay coincidencias entra la gráficas anteriores?  ¿Algún hallazgo? Comenta tu análisis.

En la primera grafica se puede observar que entidades federativas como Oaxaca, Chiapas, Campeche y Guerrero, las cuales son entidades que tienen un mayor porcentaje de municipios con grados de marginación que varían desde altos a muy altos. Uno de los casos que más llama la atención es el de Oaxaca, debido a que es una entidad federativa la cual tiene un gran número de municipios, si se contrasta con la segunda grafica (Porcentaje de población, respecto a la población total de cada estado) se puede observar que el grado de marginación aumenta conforme va disminuyendo el porcentaje de la población total con respecto al total de la entidad federativa. La entidad federativa de Oaxaca se caracteriza por tener una gran cantidad de municipios los cuales se encuentran dispersos a lo largo del territorio que comprende el estado, esta dispersión en combinación con la morfología geológica y el clima generan un gran reto para desarrollar comunicación e infraestructura logística que sea capaz de conectar a las comunidades y que por consecuencia, se generen grados de marginación tan altos.

## Grafica la relación de porcentaje de analfabetismo respecto al porcentaje de poblaciones en localidades de menos de 5,000 habitantes.

```python
plt_scatter_analfabetismo = px.scatter(x=df_marginacion_clean['% Población en localidades con menos de 5 000 habitantes'],
                         y=df_marginacion_clean['% Poblacion de 15 anos o mas analfabeta'],
                         color=df_marginacion_clean['Grado de marginación, 2020'],
                         title='porcentaje de poblacion analfabeta, respecto a la población en localidaes con menos de 5000',
                         hover_name=df_marginacion_clean['Nombre de la entidad'],
                         )

plt_scatter_analfabetismo.update_layout(xaxis_title = '% Población en localidades con menos de 5 000 habitantes',
                    yaxis_title = '% Poblacion de 15 años o mas analfabeta')
plt_scatter_analfabetismo.show()
```

## ¿Existe una relación? ¿Cómo podrías analizar con que variable tiene más corelación el porcentaje de analfabetismo en personas mayores de 15 años?

Se observa una relación muy fuerte entre el porcentaje de población analfabeta y poblaciones con menos de 5000 habitantes. Prosiguiendo con el argumento anteriormente dicho, las comunidades con un difícil acceso, ya sea por cuestiones climatologías, geológicas o por falta de infraestructura logística son las más propensas a padecer falta de servicios públicos. En el caso del analfabetismo, se da por falta de infraestructura educativa y de fuerza laboral en el ámbito educativo, en el caso de la fuerza laboral, es de necesidad traer desde afuera a personal educativo, sin embargo, la oportunidad no es atractiva debido a la lejanía de la zona y, en algunos casos, la nula existencia de servicios como internet, energía eléctrica y salud pública. Una variable la cual considero que tiene una fuerte correlación con el porcentaje de analfabetismo es la de porcentaje de población sin educación básica. Otra variable que considero tiene gran correlación con la marginación en general es el % de infraestructura logística, la cual no está considerada en los datos proporcionados.

### MIGRACION

```python
df_migracion.head()
```

```python
df_marginacion_migracion = df_marginacion_clean[['Nombre del municipio', '% Población ocupada con ingresos menores a 2 salarios mínimos']]
df_marginacion_migracion = df_marginacion_migracion.sort_values('Nombre del municipio')
df_marginacion_migracion
```

```python
df_migracion.sort_values('NOM_MUN')
```

```python
df_marginacion_migracion['Viviendas con emigrantes con destino a Esatdos Unidos'] = df_migracion['VIV_EMIG']
df_marginacion_migracion['Grado del Índice de Intensidad Migratoria'] = df_migracion['GIM-DP2']
```

```python
df_marginacion_migracion['% sin educacion basica'] = df_marginacion_clean['% Población  de 15 años o más sin educación básica']
```

```python
df_marginacion_migracion
```

```python
plt_scatter_migracion = px.scatter(x=df_marginacion_migracion['Viviendas con emigrantes con destino a Esatdos Unidos'],
                         y=df_marginacion_migracion['% sin educacion basica'],
                         color=df_marginacion_migracion['Grado del Índice de Intensidad Migratoria'],
                         title='porcentaje de la población sin educacion basica, respecto a viviendas con emigrantes con destino a EUA',
                         hover_name=df_marginacion_migracion['Nombre del municipio'],
                         )

plt_scatter_migracion.update_layout(xaxis_title = 'Viviendas con emigrantes con destino a Esatdos Unidos',
                    yaxis_title = '% sin educacion basica')
plt_scatter_migracion.show()
```

```python
plt_scatter_ingresos_emigrantes = px.scatter(x=df_marginacion_migracion['Viviendas con emigrantes con destino a Esatdos Unidos'],
                         y=df_marginacion_migracion['% Población ocupada con ingresos menores a 2 salarios mínimos'],
                         color=df_marginacion_migracion['Grado del Índice de Intensidad Migratoria'],
                         title='% Población ocupada con ingresos menores a 2 salarios mínimos, respecto a viviendas con emigrantes con destino a Esatdos Unidos',
                         hover_name=df_marginacion_migracion['Nombre del municipio'],
                         )

plt_scatter_ingresos_emigrantes.update_layout(xaxis_title = 'Viviendas con emigrantes con destino a Esatdos Unidos',
                    yaxis_title = '% Población ocupada con ingresos menores a 2 salarios mínimos')
plt_scatter_ingresos_emigrantes.show()
```

Considero que el data frame ilustrado en la parte inferior puede ser un buen referente al momento de desarrollar políticas públicas con enfoque a la migración, al observar la gráfica anteriormente expuesta se observa una correlación significativa en la población ocupada con ingresos menores a 2 salarios mínimos con respecto a las viviendas con emigrantes con destino a EUA.
